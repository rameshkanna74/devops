Basic vim config

nano ~/.vimrc

```
" ============================================================================
" Vim Configuration - DevOps / SRE Edition
" ============================================================================

" ===== General ==============================================================
set nocompatible
syntax on
filetype plugin indent on

set number
set relativenumber
set cursorline
set ruler
set showcmd
set showmatch

" ===== Indentation ==========================================================
set tabstop=4
set shiftwidth=4
set expandtab
set smarttab
set autoindent
set smartindent

augroup DevOpsIndentation
autocmd!
autocmd FileType yaml setlocal ts=2 sts=2 sw=2 expandtab
autocmd FileType python setlocal ts=4 sts=4 sw=4 expandtab
autocmd FileType sh,bash,zsh setlocal ts=4 sts=4 sw=4 expandtab
augroup END

" ===== Searching ============================================================
set hlsearch
set incsearch
set ignorecase
set smartcase

" ===== Display ==============================================================
set nowrap
set scrolloff=5
set wildmenu
set laststatus=2
set hidden

" Show tabs/trailing spaces
set list
set listchars=tab:>-,trail:┬╖

" Column marker (useful for YAML, Terraform, CI files)
set colorcolumn=81

" ===== Clipboard ============================================================
set clipboard=unnamedplus

" ===== Mouse ================================================================
set mouse=a

" ===== Persistent Undo ======================================================
set undofile
set undodir=~/.vim/undodir

" ===== Colors ===============================================================
set termguicolors
set background=dark

" ===== Bells =================================================================
set visualbell

" ===== Backups ==============================================================
set nobackup
set nowritebackup

" ===== Splits ===============================================================
set splitbelow
set splitright

" ===== File Searching =======================================================
set path+=**

" ===== Paste Mode ===========================================================
set pastetoggle=<F2>

" ============================================================================
" Leader Key
" ============================================================================

let mapleader = ","

" ============================================================================
" File Type Detection
" ============================================================================

augroup DevOpsFiletypes
autocmd!
autocmd BufRead,BufNewFile Dockerfile set filetype=dockerfile
autocmd BufRead,BufNewFile Jenkinsfile set filetype=groovy
autocmd BufRead,BufNewFile *.tf,*.tfvars set filetype=terraform
augroup END

" ============================================================================
" Quick Runners
" ============================================================================

augroup DevOpsRunners
autocmd!

autocmd FileType python nnoremap <buffer> <F5> :w<CR>:!python3 %<CR>

autocmd FileType sh nnoremap <buffer> <F5> :w<CR>:!bash %<CR>

autocmd FileType terraform nnoremap <buffer> <F6> :w<CR>:!terraform fmt %<CR>

augroup END

" ============================================================================
" WSL Clipboard Support
" ============================================================================

let s:clip = '/mnt/c/Windows/System32/clip.exe'

if executable(s:clip)
augroup WSLYank
autocmd!
autocmd TextYankPost * if v:event.operator ==# 'y' |
\ call system(s:clip, @0) |
\ endif
augroup END
endif

colorscheme habamax
```

Ctrl + B, " ==> horizontal split

Ctrl + B, n ==> next window
Ctrl + B, x ==> pane delete
Ctrl + B, q ==> pane details
